```python
!pip install gradio gtts pandas

import gradio as gr
import pandas as pd
import random
import os
import tempfile
from gtts import gTTS

# Load the irregular verbs data
data_url = "https://github.com/Hansukson/Application2/raw/main/irregular_verbs%20(1).csv"
verbs_df = pd.read_csv(data_url)

# Extract verbs data into a dictionary
verbs_data = verbs_df.set_index("present")[['past', 'p.p']].to_dict(orient="index")

# Expanded Compliments List (Correct Answers)
compliments = [
    "Fantastic job, {name}!",
    "Excellent work, {name}!",
    "You're doing great, {name}!",
    "Keep it up, {name}!",
    "Superb, {name}!",
    "Outstanding, {name}! You're unstoppable!",
    "Brilliant effort, {name}!",
    "You're crushing it, {name}!",
    "Amazing work, {name}! Keep shining!"
]

# Expanded Feedback List (Wrong Answers)
wrong_feedback = [
    "Oops! That's not quite right, {name}. Give it another shot!",
    "Not there yet, {name}, but you're learning!",
    "Almost, {name}! You can do this!",
    "Don't give up, {name}! The correct answer is here for you.",
    "Keep trying, {name}! Practice makes perfect.",
    "Nice try, {name}, but let's work on this one together!",
    "Mistakes are part of learning, {name}. You'll get it next time!",
    "You're on the right track, {name}, just a little more practice!",
    "A small bump, {name}. You're still making progress!",
    "No worries, {name}. Learning takes time – you're doing great!"
]

# 음성 생성 함수
def tts_play(verb_forms):
    """Generate speech for the verb forms: Present, Past, Past Participle."""
    tts_text = f"{verb_forms[0]}, {verb_forms[1]}, {verb_forms[2]}"
    temp_audio_file = os.path.join(tempfile.gettempdir(), "verb_audio.mp3")
    tts = gTTS(tts_text)
    tts.save(temp_audio_file)
    return temp_audio_file

# 초기 변수 설정
current_verb = ""
correct_count = 0
attempt_count = 0

# 게임 시작 함수
def start_game(name):
    """Welcome message when the game starts."""
    return f"Welcome, {name}! Click 'Show me a Verb' to begin.", gr.update(visible=True)

# 동사 보여주기
def show_random_verb():
    """Select a random present tense verb."""
    global current_verb
    current_verb = random.choice(list(verbs_data.keys()))
    return current_verb

# 정답 확인 함수
def check_answer(name, user_past, user_past_participle):
    """Check if the student's answer is correct or not."""
    global correct_count, attempt_count, current_verb

    attempt_count += 1
    correct_past = verbs_data[current_verb]['past']
    correct_pp = verbs_data[current_verb]['p.p']
    audio_file = tts_play([current_verb, correct_past, correct_pp])

    # 정답 확인
    if user_past.lower() == correct_past.lower() and user_past_participle.lower() == correct_pp.lower():
        correct_count += 1
        feedback = random.choice(compliments).format(name=name)
        feedback += f" The correct forms are: {correct_past}, {correct_pp}."
    else:
        feedback = random.choice(wrong_feedback).format(name=name)
        feedback += f" The correct forms are: {correct_past}, {correct_pp}."

    score = f"Your Score: {correct_count} / {attempt_count}"
    return feedback, score, audio_file

# 입력 리셋
def reset_inputs():
    """Reset inputs for next round."""
    return "", "", "", ""

# Gradio Interface
def verb_game():
    with gr.Blocks() as app:
        gr.Markdown("# VerbMaster: Learn Irregular Verbs! 🎯")
        gr.Markdown(""" This app helps you practice the Past and Past Participle forms of irregular verbs.
        """)
        name_input = gr.Textbox(label="Enter Your Name")
        start_button = gr.Button("Start")
        welcome_output = gr.Textbox(label="Welcome Message", interactive=False)
        show_verb_button = gr.Button("Show me a Verb", visible=False)

        # Start 버튼 클릭 시
        start_button.click(start_game, inputs=name_input, outputs=[welcome_output, show_verb_button])

        gr.Markdown("---")
        present_verb_output = gr.Textbox(label="Present Verb", interactive=False)
        user_past_input = gr.Textbox(label="Enter Past Form")
        user_pp_input = gr.Textbox(label="Enter Past Participle")
        submit_button = gr.Button("Submit")

        feedback_output = gr.Textbox(label="Feedback", interactive=False)
        score_output = gr.Textbox(label="Score", interactive=False)
        tts_output = gr.Audio(label="Audio Feedback")
        continue_button = gr.Button("If you want to continue, click here", visible=False)

        # 동사 보여주기
        show_verb_button.click(show_random_verb, outputs=present_verb_output)

        # 정답 확인 및 음성 출력, Continue 버튼 활성화
        submit_button.click(
            check_answer,
            inputs=[name_input, user_past_input, user_pp_input],
            outputs=[feedback_output, score_output, tts_output]
        )
        submit_button.click(lambda: gr.update(visible=True), outputs=continue_button)

        # Continue 버튼: 입력 리셋 및 새로운 동사 보여주기
        continue_button.click(reset_inputs, outputs=[user_past_input, user_pp_input, feedback_output, present_verb_output])
        continue_button.click(show_random_verb, outputs=present_verb_output)

    return app

# Launch the app
app = verb_game()
app.launch()
```

```python
!pip install gradio gtts pandas

import gradio as gr
import pandas as pd
import random
import os
import tempfile
from gtts import gTTS

# Load the irregular verbs data
data_url = "https://github.com/Hansukson/Application2/raw/main/irregular_verbs%20(1).csv"
verbs_df = pd.read_csv(data_url)

# Extract verbs data into a dictionary
verbs_data = verbs_df.set_index("present")[['past', 'p.p']].to_dict(orient="index")

# Correct Feedback List
correct_feedback = [
    "Correct! Fantastic job, {name}!",
    "Correct! Excellent work, {name}!",
    "Correct! You're doing great, {name}!",
    "Correct! Keep it up, {name} – you're unstoppable!",
    "Correct! Outstanding effort, {name}!",
    "Correct! Brilliant work, {name} – you nailed it!",
    "Correct! Amazing job, {name}! Keep shining!",
    "Correct! You're crushing it, {name}! Well done!",
    "Correct! Superb, {name}! You’re on fire!",
    "Correct! Perfect answer, {name}! Keep going strong!"
]

# Wrong Feedback List
wrong_feedback = [
    "Wrong! Nice try, {name}. The correct forms are: {correct_past}, {correct_pp}.",
    "Wrong! Almost there, {name}. The correct answer is: {correct_past}, {correct_pp}.",
    "Wrong! Don’t give up, {name}. The correct forms are: {correct_past}, {correct_pp}.",
    "Wrong! Good effort, {name}. The correct answer is: {correct_past}, {correct_pp}.",
    "Wrong! Mistakes happen, {name}. The correct forms are: {correct_past}, {correct_pp}.",
]

# Final Encouragement Feedback
final_encouragement = [
    "You did an amazing job today, {name}! Keep it up!",
    "Well done, {name}! Your effort is truly impressive.",
    "Great work, {name}! Every mistake is a step forward.",
    "I'm proud of your hard work, {name}. You're making progress!",
    "Fantastic effort, {name}! You’re one step closer to mastering this!"
]

def tts_play(verb_forms):
    tts_text = f"{verb_forms[0]}, {verb_forms[1]}, {verb_forms[2]}"
    temp_audio_file = os.path.join(tempfile.gettempdir(), "verb_audio.mp3")
    tts = gTTS(tts_text)
    tts.save(temp_audio_file)
    return temp_audio_file

current_verb = ""
correct_count = 0
attempt_count = 0

def start_game(name):
    if name.strip() == "":
        return "Please enter your name to proceed!", gr.update(visible=False)
    # 시작 후 SHOW ME A VERB 버튼 활성화
    return f"Welcome, {name}!", gr.update(visible=True)

def show_random_verb():
    global current_verb
    current_verb = random.choice(list(verbs_data.keys()))
    # 동사 보여준 후 SUBMIT 버튼 표시
    return current_verb, gr.update(visible=True)

def check_answer(name, user_past, user_past_participle):
    global correct_count, attempt_count, current_verb
    attempt_count += 1
    correct_past = verbs_data[current_verb]['past']
    correct_pp = verbs_data[current_verb]['p.p']
    audio_file = tts_play([current_verb, correct_past, correct_pp])

    if user_past.lower() == correct_past.lower() and user_past_participle.lower() == correct_pp.lower():
        correct_count += 1
        feedback = random.choice(correct_feedback).format(name=name)
    else:
        feedback = random.choice(wrong_feedback).format(name=name, correct_past=correct_past, correct_pp=correct_pp)

    score = f"Your Score: {correct_count} / {attempt_count}"
    # 정답 확인 후 발음 버튼, continue, end 버튼 표시
    return feedback, score, audio_file, gr.update(visible=True), gr.update(visible=True), gr.update(visible=True)

def final_feedback(name):
    return f"### THE END\n{random.choice(final_encouragement).format(name=name)}"

def reset_inputs():
    return "", "", ""

def verb_game():
    with gr.Blocks() as app:
        gr.Markdown("# VerbMaster: Learn Irregular Verbs! 🎯")

        name_input = gr.Textbox(label="Your Name")
        start_button = gr.Button("START")
        welcome_output = gr.Markdown()

        show_verb_button = gr.Button("SHOW ME A VERB", visible=False)
        submit_button = gr.Button("SUBMIT", visible=False)

        present_verb_output = gr.Textbox(label="Present Verb", interactive=False)
        user_past_input = gr.Textbox(label="Enter Past Form")
        user_pp_input = gr.Textbox(label="Enter Past Participle")

        feedback_output = gr.Textbox(label="Feedback", interactive=False)
        score_output = gr.Textbox(label="Score", interactive=False)

        audio_button = gr.Button("NOT SURE HOW TO SAY IT? HEAR IT HERE!", visible=False)
        tts_output = gr.Audio(label="Audio Feedback", visible=False)

        continue_button = gr.Button("IF YOU WANT TO CONTINUE, CLICK HERE!", visible=False)
        end_button = gr.Button("IF YOU WANT TO END THIS APP, CLICK HERE!", visible=False)

        final_feedback_output = gr.Markdown(visible=False)

        # START 버튼 클릭 후 SHOW ME A VERB 활성화
        start_button.click(start_game, inputs=name_input, outputs=[welcome_output, show_verb_button])

        # SHOW ME A VERB 클릭 후 동사 표시 + SUBMIT 버튼 표시
        show_verb_button.click(show_random_verb, outputs=[present_verb_output, submit_button])

        # SUBMIT 클릭 -> 피드백, 점수, 오디오파일 생성 + 발음 버튼, continue, end 버튼 표시
        submit_button.click(
            check_answer,
            inputs=[name_input, user_past_input, user_pp_input],
            outputs=[feedback_output, score_output, tts_output, audio_button, continue_button, end_button]
        )

        # 발음 버튼 클릭 -> 오디오 표시
        audio_button.click(lambda: gr.update(visible=True), outputs=tts_output)

        # Continue 버튼 클릭 -> 입력 리셋, 새로운 동사 표시(다음 라운드)
        continue_button.click(reset_inputs, outputs=[user_past_input, user_pp_input, feedback_output])
        continue_button.click(show_random_verb, outputs=[present_verb_output, submit_button])

        # End 버튼 클릭 시 최종 피드백 표시
        end_button.click(final_feedback, inputs=name_input, outputs=final_feedback_output)
        end_button.click(lambda: gr.update(visible=True), outputs=final_feedback_output)

    return app

app = verb_game()
app.launch()
```
